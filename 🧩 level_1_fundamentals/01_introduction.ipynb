## üß† 1. What is Python?
**Python** is a high-level, interpreted programming language known for its simplicity, readability, and versatility.

- Created by **Guido van Rossum** in **1991**
- Emphasizes **code readability** and **developer productivity**
- Used in **web development**, **data science**, **AI/ML**, **automation**, and **cybersecurity**



## ‚öôÔ∏è 2. Features of Python
| Feature | Description |
|----------|-------------|
| **Easy to learn** | Simple, English-like syntax |
| **Interpreted** | No need for compilation |
| **Object-Oriented** | Everything is an object |
| **Extensive Libraries** | Huge ecosystem (NumPy, Flask, TensorFlow, etc.) |
| **Cross-platform** | Runs on Windows, macOS, Linux |
| **Dynamic Typing** | No need to declare variable types |
