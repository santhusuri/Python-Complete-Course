## ðŸ§  1. Introduction

**Operators** are special symbols in Python that perform operations on values or variables.

Example:
```python
x = 5
y = 3
print(x + y)  # '+' is an operator
```



## ðŸ”¹ 2. Types of Operators in Python

| Category | Description |
|-----------|-------------|
| Arithmetic | Mathematical operations |
| Assignment | Assign values to variables |
| Comparison | Compare two values |
| Logical | Combine conditional statements |
| Bitwise | Perform bit-level operations |
| Identity | Compare memory locations |
| Membership | Test for membership in sequences |


## âž• 3. Arithmetic Operators

````
# Arithmetic Operators Example
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponent:", a ** b)
````

## ðŸ§¾ 4. Assignment Operators

````
x = 5
print("Initial x:", x)

x += 3  # x = x + 3
print("After += :", x)

x -= 2
print("After -= :", x)

x *= 4
print("After *= :", x)

x /= 3
print("After /= :", x)

x %= 2
print("After %= :", x)
````

## âš–ï¸ 5. Comparison Operators

````
a = 10
b = 20

print("a == b:", a == b)
print("a != b:", a != b)
print("a > b:", a > b)
print("a < b:", a < b)
print("a >= b:", a >= b)
print("a <= b:", a <= b)
````

## ðŸ§  6. Logical Operators

````
x = True
y = False

print("x and y:", x and y)
print("x or y:", x or y)
print("not x:", not x)
````

## ðŸ’¡ 7. Bitwise Operators

````
a = 5   # 0101 in binary
b = 3   # 0011 in binary

print("a & b (AND):", a & b)
print("a | b (OR):", a | b)
print("a ^ b (XOR):", a ^ b)
print("~a (NOT):", ~a)
print("a << 1 (Left Shift):", a << 1)
print("a >> 1 (Right Shift):", a >> 1)
````

## ðŸªž 8. Identity Operators

````
x = [1, 2, 3]
y = [1, 2, 3]
z = x

print("x is y:", x is y)        # False, different objects
print("x is z:", x is z)        # True, same object
print("x is not y:", x is not y)
````

## ðŸ” 9. Membership Operators

````
fruits = ["apple", "banana", "cherry"]

print("'apple' in fruits:", "apple" in fruits)
print("'mango' not in fruits:", "mango" not in fruits)
````


## ðŸ§© 10. Operator Precedence

Operator precedence defines the order in which operations are evaluated.

Example:
```python
x = 10 + 2 * 3
print(x)  # 16 because * has higher precedence than +
```

| Operator | Description |
|-----------|-------------|
| `()` | Parentheses |
| `**` | Exponentiation |
| `*`, `/`, `//`, `%` | Multiplication, Division, etc. |
| `+`, `-` | Addition, Subtraction |
| `==`, `!=`, `>`, `<` | Comparisons |
| `and`, `or`, `not` | Logical operators |


````
x = 10 + 2 * 3
y = (10 + 2) * 3
print("Without parentheses:", x)
print("With parentheses:", y)
````
