## ðŸ§  1. Introduction

**Conditional statements** allow your program to make decisions based on certain conditions.  
They help control the flow of execution â€” only executing blocks of code when conditions are met.

Common keywords:
- `if`
- `elif`
- `else`



## ðŸ§© 2. Basic `if` Statement

Syntax:
```python
if condition:
    # Code to execute if condition is True
```

Example:


````
x = 10
if x > 5:
    print("x is greater than 5")
````

## ðŸ”¹ 3. `if-else` Statement

````
x = 3
if x >= 5:
    print("x is greater than or equal to 5")
else:
    print("x is less than 5")
````

## ðŸ”¸ 4. `if-elif-else` Ladder

````
marks = 82

if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 60:
    print("Grade: C")
else:
    print("Grade: Fail")
````

## ðŸ§± 5. Nested `if` Statements

````
age = 18
citizen = True

if age >= 18:
    if citizen:
        print("You are eligible to vote!")
    else:
        print("You must be a citizen to vote.")
else:
    print("You must be at least 18 years old.")
````

## âš™ï¸ 6. Logical Operators with Conditions

````
age = 25
has_license = True

if age >= 18 and has_license:
    print("You can drive.")
else:
    print("You cannot drive.")
````

## âœ³ï¸ 7. Ternary (Inline) If-Else

````
num = 10
result = "Even" if num % 2 == 0 else "Odd"
print(f"{num} is {result}")
````


## ðŸ§© 8. Real-World Example â€” Temperature Check


````
temperature = 32

if temperature > 30:
    print("It's a hot day!")
elif temperature >= 20:
    print("It's a warm day!")
else:
    print("It's cold outside.")
````


## ðŸ§® 9. Beginner-Level Challenges

### 1ï¸âƒ£ Check if a number is even or odd
### 2ï¸âƒ£ Determine if a number is positive, negative, or zero

````
# 1ï¸âƒ£ Even or Odd
num = 7
if num % 2 == 0:
    print(f"{num} is even.")
else:
    print(f"{num} is odd.")
````

````
# 2ï¸âƒ£ Positive, Negative, or Zero
num = -5
if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
````

````
# 3ï¸âƒ£ Grade Calculator
marks = 68
if marks >= 90:
    grade = "A"
elif marks >= 75:
    grade = "B"
elif marks >= 60:
    grade = "C"
else:
    grade = "Fail"
print(f"Marks: {marks} | Grade: {grade}")
````


## ðŸ’ª 10. Advanced Challenges

### 1ï¸âƒ£ Find the largest of three numbers
### 2ï¸âƒ£ Check if a year is a leap year
### 3ï¸âƒ£ Simple discount calculator


````
# 1ï¸âƒ£ Largest of Three Numbers
a, b, c = 15, 22, 9

if a >= b and a >= c:
    largest = a
elif b >= a and b >= c:
    largest = b
else:
    largest = c

print(f"Largest number: {largest}")
````

````
# 2ï¸âƒ£ Leap Year Checker
year = 2024

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year!")
else:
    print(f"{year} is not a leap year.")
````

````
# 3ï¸âƒ£ Discount Calculator
amount = 1200

if amount >= 1000:
    discount = 0.15
elif amount >= 500:
    discount = 0.10
else:
    discount = 0.05

final_amount = amount - (amount * discount)
print(f"Bill: â‚¹{amount}, Discount: {discount*100}%, Final: â‚¹{final_amount}")
````


## ðŸ§  Summary

| Nested if | One `if` inside another |
| Logical Operators | `and`, `or`, `not` |
| Ternary | Inline one-line condition |
