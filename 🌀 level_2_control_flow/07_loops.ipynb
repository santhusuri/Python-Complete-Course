## üß† 1. Introduction

**Loops** in Python are used to execute a block of code repeatedly until a condition is met.

Types of loops:
- `for` loop ‚Üí iterates over a sequence (like a list, string, or range)
- `while` loop ‚Üí repeats as long as a condition is true



## üîÅ 2. `for` Loop

Syntax:
```python
for variable in sequence:
    # code block
```

Example:


````
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print("I like", fruit)
````

## üßÆ 3. Using `range()` Function

````
for i in range(5):
    print("Number:", i)
````

````
# Range with start and step
for i in range(1, 11, 2):
    print(i, end=" ")
````

## üîÑ 4. `while` Loop

````
count = 0
while count < 5:
    print("Count:", count)
    count += 1
````

## üõë 5. Loop Control Statements


| Statement | Description |
|------------|-------------|
| `break` | Exits the loop prematurely |
| `continue` | Skips the rest of the current iteration |
| `pass` | Placeholder statement (does nothing) |


````
# Example of break
for i in range(10):
    if i == 5:
        break
    print(i)
````

````
# Example of continue
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
````

````
# Example of pass
for i in range(3):
    pass  # Placeholder
print("Loop executed successfully!")
````

## üß© 6. `else` Block with Loops

````
for num in range(3):
    print(num)
else:
    print("Loop completed successfully!")
````

````
# Using break cancels the else block
for num in range(3):
    if num == 1:
        break
    print(num)
else:
    print("This will not print")
````


## üßÆ 7. Nested Loops


````
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i} x {j} = {i*j}")
    print("---")
````


## üí™ 9. Advanced Challenges

### 1Ô∏è‚É£ Reverse a number  
### 2Ô∏è‚É£ Check if a number is prime  
### 3Ô∏è‚É£ Print a pattern (triangle or pyramid)  
### 4Ô∏è‚É£ Find factorial of a number using while loop


````
# 1Ô∏è‚É£ Reverse a number
num = 12345
rev = 0
while num > 0:
    rev = (rev * 10) + (num % 10)
    num //= 10
print("Reversed Number:", rev)
````

````
# 2Ô∏è‚É£ Prime number check
num = 17
is_prime = True

if num <= 1:
    is_prime = False
else:
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

print(f"{num} is prime:", is_prime)
````

````
# 3Ô∏è‚É£ Print pyramid pattern
rows = 5
for i in range(1, rows + 1):
    print("*" * i)
````

````
# 4Ô∏è‚É£ Factorial using while loop
num = 5
fact = 1
while num > 0:
    fact *= num
    num -= 1
print("Factorial:", fact)
````


## üß† Summary

| Concept | Description |
|----------|-------------|
| `for` loop | Iterates over a sequence |
| `while` loop | Runs until a condition becomes False |
| `break` | Exits loop prematurely |
| `continue` | Skips current iteration |
| `else` with loop | Executes if loop completes without break |
| Nested loops | Loop inside another loop |
